Python Functions & Iterators - Theory Questions
1. Difference between a Function and a Method in Python
A function is an independent block of code, while a method is a function defined inside a class.

Example:
def add(a, b):
    return a + b

class Calculator:
    def multiply(self, a, b):
        return a * b

print(add(2, 3))
print(Calculator().multiply(2, 3))





2. Function Arguments and Parameters
Parameters are placeholders in function definition, arguments are values passed during call.

Example:
def greet(name):
    print('Hello', name)

greet('Mohit')




3. Ways to Define and Call a Function
Ways include: normal function, lambda, recursive function.

Example:
def square(x): return x*x
print(square(5))

double = lambda x: x*2
print(double(10))

def fact(n): return 1 if n==0 else n*fact(n-1)
print(fact(5))





4. Purpose of return Statement
It returns a value from a function.

Example:
def add(a, b):
    return a + b

print(add(5, 7))






5. Iterators vs Iterables
Iterable: can be looped (list, tuple, str). Iterator: produces next item using __next__().

Example:
nums = [1,2,3]
it = iter(nums)
print(next(it))
print(next(it))





6. Generators
Generators use 'yield' to produce values lazily.

Example:
def gen_numbers():
    for i in range(3):
        yield i

for x in gen_numbers():
    print(x)






7. Advantages of Generators
Memory efficient and simpler iterator implementation.

Example:
nums = (i for i in range(1000000))





8. Lambda Functions
Anonymous small functions used for short tasks.

Example:
square = lambda x: x*x
print(square(6))





9. map() Function
Applies a function to each element.

Example:
nums = [1,2,3,4]
print(list(map(lambda x: x*x, nums)))





10. Difference between map(), reduce(), filter()
map() → transforms, filter() → selects, reduce() → accumulates.

Example:
from functools import reduce
nums = [1,2,3,4,5]
print(list(map(lambda x: x*2, nums)))
print(list(filter(lambda x: x%2==0, nums)))
print(reduce(lambda x,y: x+y, nums))
